# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv(r'C:\Users\OneDrive\Desktop\New folder\glassdoor_jobs.csv')
```

```python
# Display basic information
print("Dataset Overview:")
df.info()
```

```python
print("\nFirst few rows:")
print(df.head())
```

```python
# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
# Unique values in categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nUnique values in {col}:")
    print(df[col].unique()[:10])  # Show first 10 unique values
```

```python
# Salary Data Cleaning
df = df[df['Salary Estimate'].notna()]
df['Salary Estimate'] = df['Salary Estimate'].astype(str)
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split(' ')[0] if 'Glassdoor' in x else x)
df['Salary Estimate'] = df['Salary Estimate'].str.replace('[^0-9K-]', '', regex=True)
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('-') if '-' in x else [x, x])
df = df[df['Salary Estimate'].apply(lambda x: len(x) == 2)]  # Ensure valid ranges

def safe_int_conversion(val):
    try:
        return int(val.replace('K', ''))
    except ValueError:
        return np.nan

df['Min Salary'] = df['Salary Estimate'].apply(lambda x: safe_int_conversion(x[0]))
df['Max Salary'] = df['Salary Estimate'].apply(lambda x: safe_int_conversion(x[1]))
df.dropna(subset=['Min Salary', 'Max Salary'], inplace=True)
df['Avg Salary'] = (df['Min Salary'] + df['Max Salary']) / 2
```

```python
# 1. Distribution of Salaries
plt.figure(figsize=(10, 5))
sns.histplot(df['Avg Salary'], bins=30, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Average Salary (in K)')
plt.ylabel('Frequency')
plt.show()
```

```python
#2.Countplot of Job Titles
plt.figure(figsize=(12, 6))
sns.countplot(y=df['Job Title'], order=df['Job Title'].value_counts().index[:10])
plt.title('Top 10 Job Titles')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()
```

```python
# 3. Company Ratings Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Rating'].dropna(), bins=20, kde=True)
plt.title('Company Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Industry Analysis
top_industries = df['Industry'].value_counts().head(10)
top_industries.plot(kind='bar', figsize=(10, 5))
plt.title('Top 10 Industries')
plt.ylabel('Count')
plt.xlabel('Industry')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Revenue Distribution
plt.figure(figsize=(10, 5))
df['Revenue'].value_counts().head(10).plot(kind='bar')
plt.title('Top Revenue Ranges')
plt.ylabel('Count')
plt.xlabel('Revenue Range')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Type of Ownership Distribution
plt.figure(figsize=(10, 5))
df['Type of ownership'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Type of Ownership Distribution')
plt.show()
```

```python
# 7. Size Distribution
plt.figure(figsize=(10, 5))
df['Size'].value_counts().plot(kind='bar')
plt.title('Company Size Distribution')
plt.ylabel('Count')
plt.xlabel('Size')
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Founded Year Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Founded'], bins=30, kde=True)
plt.title('Distribution of Company Founded Year')
plt.xlabel('Year Founded')
plt.ylabel('Count')
plt.show()
```

```python
# 9. Salary vs. Rating
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Rating'], y=df['Avg Salary'])
plt.title('Salary vs. Company Rating')
plt.xlabel('Company Rating')
plt.ylabel('Average Salary')
plt.show()
```

```python
# 10. Location-based Job Distribution
top_locations = df['Location'].value_counts().head(10)
top_locations.plot(kind='bar', figsize=(10, 5))
plt.title('Top 10 Job Locations')
plt.ylabel('Count')
plt.xlabel('Location')
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. Correlation Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 12. Competitors Analysis
plt.figure(figsize=(10, 5))
df['Competitors'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Competitor Mentions')
plt.ylabel('Count')
plt.xlabel('Competitors')
plt.xticks(rotation=45)
plt.show()
```

```python
# 13. Job Sector Distribution
plt.figure(figsize=(10, 5))
df['Sector'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Job Sectors')
plt.ylabel('Count')
plt.xlabel('Sector')
plt.xticks(rotation=45)
plt.show()
```

```python
# 16. Job Description Word Count Distribution
df['Job Description Length'] = df['Job Description'].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(10, 5))
sns.histplot(df['Job Description Length'], bins=30, kde=True)
plt.title('Distribution of Job Description Word Count')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
```
